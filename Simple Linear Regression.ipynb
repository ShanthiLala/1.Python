```python
#The required libraries are imported at the beginning of the program.
#Pandas is used for handling datasets.
import pandas as pd
```

```python
#Load the dataset is loaded into the dataset variable using a CSV file.
dataset = pd.read_csv("Salary_Data.csv")
```

```python
#print dataset and Understand the Dataset
dataset
```

```python
#The dataset is divided into:
#Independent Variable (x) – Input
#Dependent Variable (y) – Output
#In simple linear regression, the model tries to learn the relationship: Y = mX + c
#Where: m = slope c = intercept
#This step prepares the data for training.
# Assign input to independant Variable
independant = dataset[["YearsExperience"]]
#print independant variable
independant
```

```python
# Assign output to dependendant Variable
dependendant =dataset[["Salary"]]
#print dependendant variable
dependendant
```

```python
#The data is divided into: Training data and Testing data
#The training data is used to build the model.
#The testing data is used to evaluate model performance.
#To do this we are using sklearn
#sklearn.model_selection imports the train_test_split function so you can use it to split your data into training and testing sets.
from sklearn.model_selection import train_test_split
```

```python
#print train_test_split
train_test_split
```

```python
# The below line of code splits data into 4 parts:
# x_train = input features for training
# x_test = input features for testing
# y_train = target/output for training
# y_test = target/output for testing
# And the options mean: 
# test_size=0.30 → 30% of the data goes to test, 70% to train
# random_state=0 → keeps the split same every time you run it (for repeatable results)
#So Split independant and dependendant into train/test sets (70/30) in a repeatable way.   
x_train,x_test,y_train,y_test=train_test_split(independant, dependendant, test_size=0.30, random_state=0)
```

```python
# imports the Linear Regression model.
from sklearn.linear_model import LinearRegression
# Creates a new linear regression model object (named regressor)
regressor = LinearRegression()
# Trains the model using your training data, so it learns the best line/relationship from x_train to predict y_train.
regressor.fit(x_train, y_train)
```

```python
# Gets the learned coefficients (weights / slopes) from the trained model.
weight = regressor.coef_
# Prints those coefficient values.
weight
```

```python
# Gets the model’s intercept (bias), the value when all X = 0
bais = regressor.intercept_
# Prints that intercept value.
bais
```

```python
# Uses the trained model to predict output values for the test inputs
y_pred = regressor.predict(x_test)
# Prints those predicted values
y_pred
```

```python
# Imports the R² score function.
from sklearn.metrics import r2_score
# Calculates the R² value comparing actual (y_test) vs predicted (y_pred).
r2_score = r2_score(y_test, y_pred)
# Print r2_score
r2_score
```

```python
# loads the pickle tool, used to save and load Python objects
import pickle
# Sets the file name where you will save (or load) the trained model.
filename = "finalised_model_simple_linear.sav"
```

```python
# This saves your trained regressor model into the file finalised_model_simple_linear.sav in binary write mode ("wb").
pickle.dump(regressor, open(filename,"wb"))
```

```python
# This loads the saved model from finalised_model_simple_linear.sav in binary read mode ("rb") and stores it in loaded_model
loaded_model = pickle.load(open("finalised_model_simple_linear.sav","rb"))
```

```python
# Uses the loaded model to predict the output for input value 15.
result = loaded_model.predict([[15]])
# Prints that predicted value
result
```
